```python
# Dependencies and Setup
import pandas as pd
import numpy as np
```

```python
# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"
```

```python
# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)
```

```python
# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
#Calculate the total number of schools
schools_total = school_data_complete['school_name'].nunique()
```

```python
#Calculate the total number of students
students_total = school_data_complete['Student ID'].count()
```

```python
#Calculate the total budget
total_budget = school_data_complete['budget'].unique().sum()
```

```python
#Calculate the average math score
math_avg = school_data_complete['math_score'].mean()
```

```python
#Calculate the average reading score
read_avg = school_data_complete['reading_score'].mean()
```

```python
#Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2
overall_avg = (school_data_complete['math_score'].mean() + school_data_complete['reading_score'].mean()) / 2
```

```python
#Calculate the percentage of students with a passing math score (70 or greater)
pass_math = sum((school_data_complete['math_score'] >= 70)) / \
                (school_data_complete['math_score'].count()) * 100
```

```python
#Calculate the percentage of students with a passing reading score (70 or greater)
pass_read = sum((school_data_complete['reading_score']) >= 70) / \
                (school_data_complete['reading_score'].count()) * 100
```

```python
#Create a dataframe to hold the above results
district_summary_df = pd.DataFrame({'Total Schools': [schools_total], 
                                    'Total Students': [students_total], 
                                    'Total Budget': [total_budget], 
                                    'Average Math Score': [math_avg], 
                                    'Average Reading Score': [read_avg], 
                                    '% Passing Math': [pass_math], 
                                    '% Passing Reading': [pass_read], 
                                    'Overall Average': [overall_avg]})
district_summary_df
```
